```python
import random
from datetime import datetime

class Item:
    def __init__(self):
        self.names = []
        
    def getRandomFromList(self,L):
        i = random.randint(0,len(L)-1)
        return L[i]
    
    def setName(self):        
        self.name = self.getRandomFromList(self.names)
    
    def setMaterial(self):        
        self.material = self.getRandomFromList(self.mats)
        
    def setDecoration(self):        
        self.decoration = self.getRandomFromList(self.decorations)
        
    def setQuality(self):        
        self.quality = self.getRandomFromList(self.qualities)
    
class ItemController:    
    def __init__(self, item):
        self.item = item
        
        item = []
        
    def setItemSet(Item):
        N = 10
        for i in range(N):
            self.items = Item()
    
    def getItemDetails(self):
        details = self.item.decoration+ " " + self.item.quality + " " + self.item.material + " " + self.item.name
        return details    
    
class Weapon(Item):
    def __init__(self):        
        super().__init__()
        self.names = ["Sword", "Axe", "Bow", "Spear", "Halberd", "Scimitar", "Longsword"]
        self.mats = ["Iron", "Mithril", "Adamantite", "Starstone", "Abyssal", "Arkhite", "Steel", "Blacksteel", "Dragonsteel"]
        self.decorations = ["Runed", "Inscribed", "Gilded", "Jewelled", "Glittering", "Glowing", "Fiery", "Sparking", "Cold", "Blooded"]
        self.qualities = ["Shoddy", "Worn", "Weathered", "Used", "Regular", "Common", "Fine", "Exquisite", "Masterwork", "Divine", "Fell"]
        
    def attack(index):
        self.attacks[index].use()
        
class TargetStrategy:
    def getTarget(self):
        return self.targetAction + self.targetRange
    
class Swing(TargetStrategy):
    def __init__(self):
        super().__init__()
        self.targetAction = "swings at enemies "
        self.targetRange = "in arms reach"
class Thrust(TargetStrategy):
    def __init__(self):
        super().__init__()
        self.targetAction = "thrusts at enemies "
        self.targetRange = "a short step away"
        
class Slam(TargetStrategy):
    def __init__(self):
        super().__init__()
        self.targetAction = "slams down upon enemies"
        self.targetRange = "in front of you"

class Projectile(TargetStrategy):
    def __init__(self):
        super().__init__()
        self.targetAction = "sails towards enemies "
        self.targetRange = "in the distance"

class AoE(TargetStrategy):
    def __init__(self):
        super().__init__()
        self.targetAction = "blasts enemies"
        self.targetRange = "in a large area around you"
        
class Melee(Weapon):
    def __init__(self):
        super().__init__()
        
class Sword(Melee):
    def __init__(self):
        super().__init__()

class Sword2H(Sword):
    def __init__(self):
        super().__init__()
        self.names = ["Greatsword", "Bastard Sword", "Zweihander", "Katana"]

class Sword1H(Sword):
    def __init__(self):
        super().__init__()
        self.names = ["Longsword", "Cutlass", "Falchion", "Rapier"]

class Axe(Melee):
    def __init__(self):
        super().__init__()
        self.names = ["Axe", "Hatchet"]        

class Polearm(Melee):
    def __init__(self):
        super().__init__()
        self.names = ["Spear", "Halberd", "Pike", "Lance"]
    
class ShortBlade(Melee):
    def __init__(self):
        super().__init__()
        self.names = ["Dagger", "Knife", "Hook", "Scythe", "Shortsword"]
    
class Armour(Item):
    def __init__(self):        
        super().__init__()
        self.names = ["Skirt", "Greaves", "Gloves", "Torso", "Boots", "Pauldron", "Helmet", "Shield"]
        self.mats = ["Leather", "Cloth", "Iron", "Mithril", "Adamantite", "Abyssal", "Arkhite", "Steel", "Blacksteel", "Dragonsteel"]
        self.decorations = ["Runed", "Inscribed", "Gilded", "Jewelled", "Dirty", "Dented"]
        self.qualities = ["Shoddy", "Worn", "Weathered", "Used", "Regular", "Common", "Fine", "Exquisite", "Masterwork"]
    
N = 10
armourList = []
weaponList = []

for i in range(N):
    armourList.append(ItemController(Armour()))
    armourList[i].item.setName()
    armourList[i].item.setMaterial()
    armourList[i].item.setDecoration()
    armourList[i].item.setQuality()
    
    weaponList.append(ItemController(Axe()))
    weaponList[i].item.setName()
    weaponList[i].item.setMaterial()
    weaponList[i].item.setDecoration()
    weaponList[i].item.setQuality()
    
    

print("Sword1H:")
for i in range(N):    
    print(armourList[i].getItemDetails())
    
print("\n")

print("Axe:")
for i in range(N):
    print(weaponList[i].getItemDetails())
    
    
    

```
